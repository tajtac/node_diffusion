```python
import numpy as np
```

```python
name = 'PP_fine_filleted'
with open(name + '.txt') as f:
    lines = [line.rstrip('\n') for line in f]
n_node = int(lines[1])
n_elems = int(lines[n_node+3])
node_X      = np.array([line.split() for line in lines[2:n_node+2]], dtype=float) #load all the nodes
elements    = np.array([line.split() for line in lines[n_node+4:] ], dtype=int) #load all the elements

node_X      = np.hstack([np.arange(1,n_node+1)[:,None], node_X]) #add numbering for nodes
elements    = np.hstack([np.arange(1,n_elems+1)[:,None], elements]) #add numbering for elements
```

```python
inpfile = "*Heading\n** Job name: Pmesh_ordered Model name: Pmesh_ordered\n** Generated by: Abaqus/CAE 2020 \
    \n*Preprint, echo=NO, model=NO, history=NO, contact=NO\n**\n** PARTS\n**\n*Part, name=Part-1\n*Node\n"

for x in node_X:
    inpfile+= '{}, {}, {}\n'.format(int(x[0]), x[1], x[2])

inpfile+= '*Element, type=CPS3\n'

for x in elements:
    inpfile+= '{}, {}, {}, {}\n'.format(int(x[0]), x[1], x[2], x[3])

with open(name + '.inp', 'w') as f:
    f.write(inpfile)
```
